```python
import numpy as np
data1 = np.genfromtxt('profile1.data', delimiter=' ')
data2 = np.genfromtxt('profile2.data', delimiter=' ')
```

```python
data1[:, 0] = data1[:, 0] - data1[0, 0]
data2[:, 0] = data2[:, 0] - data2[0, 0]
```

```python
x1 = data1[:, 0]
y_origin1 = data1[:, 1]
y1 = np.cumsum(y_origin1)
x2 = data2[:, 0]
y_origin2 = data2[:, 1]
y2 = np.cumsum(y_origin2)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(x1,y1)
plt.show()
plt.plot(x2,y2)
plt.show()
```
